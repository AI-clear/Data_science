[코드 3-6] CSV 파일 작성

```python
#csv 라이브러리 가져오기
import csv

#csv 파일 불러오기
f = open('write.csv','w',encoding='cp949', newline='')

# #csv 파일 작성할 준비하기
wr = csv.writer(f)

# #데이터를 한 행씩 작성하기
wr.writerow(['ID','이름','상징색','취미','특징'])
wr.writerow(['001','뽀로로','파랑색','낚시','펭귄'])
wr.writerow(['002','에디','주황색','과학실험','사막여우'])
wr.writerow(['003','크롱','초록색','눈싸움','공룡'])
wr.writerow(['004','루피','분홍색','요리','비버'])
wr.writerow(['005','패티','보라색','운동','펭귄'])


# #f 변수의 자원 반환하기
f.close()
```

[코드 3-7] CSV 파일에 데이터 추가

```python
f = open('write.csv','w',encoding='cp949', newline='')
wr = csv.writer(f)

wr.writerow(['006','포비','흰색','춤','북극곰'])

#f 변수의 자원 반환하기
f.close()
```

[코드 3-13] 새 워크시트 생성

[코드 3-14] 워크시트 이름 변경

```python
sheet1 = wb['Sheet1']
sheet2 = wb['Sheet2']

sheet1.title = '캐릭터 명단'
sheet2.title = '인기도 조사'

print(wb.sheetnames)
```

[코드 3-15] 워크시트 이름 변경

```python
sheet2['B1'] = '인기도 조사결과'
print(sheet2['B1'].value)
```

[코드 3-16] 워크시트 복제

```python
copysheet = wb.copy_worksheet(sheet2)
print(wb.sheetnames)

#복제 시트의 이름 변경하기
copysheet.title = 'copy’

print(wb.sheetnames)
print(copysheet['B1'].value)
```

[코드 3-17] 워크시트 삭제

```python
del wb['인기도 조사']
print(wb.sheetnames)
```
