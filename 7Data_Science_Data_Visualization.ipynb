노드와 선 추가 및 제거

[코드 7-13] 노드와 선 추가 및 제거

```python
#노드 추가 및 제거하기
g1 = nx.Graph()
g1.add_node('David')
g1.add_node('John')
g1.add_node('Mary')
g1.add_node('Julia')
g1.add_node('Juda')
g1.add_nodes_from(['Judy', 'Karen'])
g1.remove_node('Julia')

#선 추가 및 제거하기(연결된 선의 순서는 상관 쓰지 않음)
g1.add_edge('David', 'John')
g1.add_edge('David', 'Mary')
g1.add_edge('David', 'Karen')
g1.add_edges_from([('John', 'Mary'), ('John', 'Judy')])
g1.remove_edge('Mary', 'John')

nx.draw(g1, with_labels=True, font_weight='bold')
```

노드와 크기 차수에 비례하게 설정

[코드 7-14] 노드 크기를 차수에 비례하게 설정

```python
d = dict(g1.degree) # 그래프 g1의 각 노드의 차수를 딕셔너리형으로 저장
nx.draw(g1, nodelist=d.keys(), node_size=[v * 1000 for v in d.values()],\
with_labels=True, font_weight='bold')
g1.nodes
```

방향성 그래프

[코드 7-15] 방향성 그래프

```python
g2 = nx.DiGraph()
g2.add_edges_from([(1, 2), (1, 3), (1, 4), (3, 4)])
nx.draw(g2, with_labels=True, font_weight='bold')
print(g2.in_degree) # [(1, 0), (2, 1), (3, 1), (4, 2)]
print(g2.out_degree) # [(1, 3), (2, 0), (3, 1), (4, 0)]
```

가중치 표시

[코드 7-16] 가중치 표시

```python
g3 = nx.DiGraph()
g3.add_weighted_edges_from([(1, 2, 10), (2, 3, 20)])
g3.add_edge(1, 3, weight=30)
pos = nx.spring_layout(g3)
nx.draw(g3, pos=pos, with_labels=True)
labels = nx.get_edge_attributes(g3, 'weight')
nx.draw_networkx_edge_labels(g3, pos, edge_labels=labels);
```

[코드 7-33] 라이브러리 불러오기

```python
import numpy as np
import pandas as pd
import re
import matplotlib.pyplot as plt
import networkx as nx
from konlpy.tag import Hannanum

hannanum = Hannanum()
```

[코드 7-34] 파일을 열어 텍스트 읽기

```python
f = open('/content/drive/MyDrive/Colab Notebooks/실습데이터-파일입출력/News.csv', 'r', encoding='cp949')
lines = f.readlines()
print(lines)
f.close()
```

[코드 7-35] 텍스트에서 명사만 추출

```python
dataset = []
for i in range(len(lines)):
  dataset.append(hannanum.nouns(re.sub('[^가-힣a-zA-Z\s]','', lines[i])))
dataset[:3]
```

[코드 7-36] 연관규칙 분석


```python
!pip install apyori
from apyori import apriori

#연관규칙 데이터프레임 생성하기
result = (list(apriori(dataset, min_support=0.01)))
df = pd.DataFrame(result)
#항목 수가 2인 것만 남기기
df['length'] = df['items'].apply(lambda x: len(x))
df = df[(df['length']==2)&(df['support']>=0.01)].sort_values(by='support',\
ascending=False)
df.head(3)
```

[코드 7-37] 네트워크 그래프 생성

```python
!pip install scipy==1.8.0
!pip install pyg-nightly

G = nx.Graph()
ar = (df['items']); G.add_edges_from(ar)
pr = nx.pagerank(G)
nsize = np.array([v for v in pr.values()])
nsize = 2000 * (nsize-min(nsize)) / (max(nsize)-min(nsize))
pos = nx.circular_layout(G)
plt.figure(figsize=(16,12)); plt.axis('off')
nx.draw_networkx(G, font_family='NanumGothic', font_size=20, pos=pos,\
node_color=list(pr.values()), node_size=nsize*10, alpha=0.7,\
edge_color='.5', cmap=plt.cm.coolwarm)
plt.savefig('IMG.png', bbox_inches='tight')
```

```python
!pip install konlpy
!pip install networkx==2.7.1
import numpy as np
import pandas as pd
import re
import matplotlib.pyplot as plt
import networkx as nx
from konlpy.tag import Hannanum

hannanum = Hannanum()
```

3.한글 폰트를 설치하고 라이브러리를 불러오기

```python
!pip install konlpy
!pip install networkx==2.7.1

f = open('/content/drive/MyDrive/Colab Notebooks/실습데이터-파일입출력/우크라이나.csv', 'r', encoding='cp949')
lines = f.readlines()
f.close()

dataset = []
for i in range(len(lines)):
  dataset.append(hannanum.nouns(re.sub('[^가-힣a-zA-Z\s]','', lines[i])))
dataset[:3]
```

4.연관규칙 분석 알고리즘 apriori로 가능성이 높은 연관규칙만 계산하도록 필터링.

```python
!pip install apyori
from apyori import apriori

result = (list(apriori(dataset, min_support=0.01)))
df = pd.DataFrame(result)

df['length'] = df['items'].apply(lambda x: len(x))
df = df[(df['length']==2)&(df['support'] >= 0.01)].sort_values(by='support',
ascending=False)
df
```

5.네트워크 그래프를 생성하고 그래프에 나타난 연관관계를 분석.

```python
!pip install scipy==1.8.0
!pip install pyg-nightly

G = nx.Graph()
ar = (df['items']); G.add_edges_from(ar)
pr = nx.pagerank(G)
nsize = np.array([v for v in pr.values()])
nsize = 2000*(nsize-min(nsize)) / (max(nsize)-min(nsize))
pos = nx.spring_layout(G) #spring 레이아웃을 사용하기
plt.figure(figsize=(16,12)); plt.axis('off')
nx.draw_networkx(G, font_family='NanumGothic', font_size=20, pos=pos,\
node_color=list(pr.values()), node_size=nsize*10, alpha=0.7,\
edge_color='.5', cmap=plt.cm.coolwarm)
plt.savefig('네트워크 그래프_우크라이나.png', bbox_inches='tight')
```
